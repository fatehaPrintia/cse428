```python
import numpy as np
from skimage import io,data,exposure
import matplotlib.pyplot as plt
```

```python
import cv2
import numpy as np
from PIL import Image
import io

def add_fog(image, fog_intensity=0.5):
    # Convert the image to a NumPy array
    img = np.array(image)

    # Generate random fog effect
    fog = np.random.randint(0, 256, img.shape, dtype=np.uint8)

    # Combine the original image and the fog effect
    foggy_img = cv2.addWeighted(img, 1 - fog_intensity, fog, fog_intensity, 0)

    # Convert the NumPy array back to PIL Image
    foggy_img_pil = Image.fromarray(foggy_img)

    return foggy_img_pil

if __name__ == "__main__":
    input_image_path = "car.jpg"
    foggy_intensity = 0.5  # Adjust this value (0.0 to 1.0) to control fog intensity

    # Load the image using Pillow (PIL)
    with open(input_image_path, 'rb') as f:
        image = Image.open(io.BytesIO(f.read()))
        image.show()

    output_image = add_fog(image, foggy_intensity)

    # Display the foggy image
    output_image.show()
    print(output_image.show())
```
